#### Implementation: Reprioritization Algorithm

4. Calculate co-failure distribution

    - add all tests of change req 0 to dispatch queue with conditional failure rate as 0.5/1
    - in change req 1, test 1 comes
        - check if the test already exists in dispatch queue
        - if no, ie it is a new test, then set its conditional failure rate to 0.5/1 and add to dispatch queue
        - if yes, find the previous test status
            - if previous test status= failed
                - new score= previous priority+ (p(current and previous test fails)- 0.5)
            - if previous test status= pass
                - new score= previous priority+ (p(current test fails and previous test pass)-0.5)

```python
# define a dataset to store frequent co failing test suites
Cofailure_Distribution = pd.DataFrame(columns = ["Test Pair","Fail Together","Run Together","Pass-Fail"])
```

```python
import traceback
#currentTest - test to be reprioritized
#t_finish - historial tests with same change request( To calculate co-fail data)
#finishedGpost - historical completed tests (To calculate the new priority)

def reprioritize(currentTest,index,t_finish,finishedGpost,dispatch_queue): 
    global Cofailure_Distribution
    test_names = [currentTest.Test_Suite,t_finish.Test_Suite]
    sorted_names = sorted(test_names)
    joined_names = ','.join(sorted_names)
    
    # get all change request numbers
    change_requests= googlepostcommit.Change_Request.unique() 

    # check if currentTest was executed before in the historical queue
    if (currentTest.Test_Suite in finishedGpost.Test_Suite.values):
        # initialise the co-fail count
        co_fail_count=0
        # initialise current fail , t_finish pass count
        co_fail_pass_count=0
        # initialise co_occurance count
        co_occurance_count=0

    
        # get the previous status
        previous_status=finishedGpost.iat[finishedGpost.Test_Suite.eq(currentTest.Test_Suite).idxmax(),3]
        #get the previous priority off currentTest
        previous_priority=finishedGpost.iat[finishedGpost.Test_Suite.eq(currentTest.Test_Suite).idxmax(),10] # priority column no=10
        
        # get all change requests before the current test change request .
        # Generate co fail/co occurance data of currentTest with respect to other tests in these change requests.
        previous_change_requests= sorted(i for i in change_requests if i < currentTest.Change_Request)
        
        # COFAILURE DISTRIBUTION USING CONDITIONAL PROBABILITY
        # iterate through the previous change requests
        for i in range(0, len(previous_change_requests)):
            prev_req = previous_change_requests[i]
            overlapping_row_count = googlepostcommit[googlepostcommit['Change_Request']<prev_req].shape[0]
#        for pre_req in previous_change_requests:
        
            # get all test suites in current change request being iterated
            test_suites_in_change_req=googlepostcommit[googlepostcommit['Change_Request']==prev_req]
            try:
                # check for co-occurance
                if(currentTest.Test_Suite in test_suites_in_change_req.Test_Suite.values and t_finish.Test_Suite in test_suites_in_change_req.Test_Suite.values):
                    # if yes, increment the co-occurance count
                    co_occurance_count+=1
                    test_suite_actual_index_currentTest = test_suites_in_change_req.Test_Suite.eq(currentTest.Test_Suite).idxmax()
                    test_suite_actual_index_tFinish = test_suites_in_change_req.Test_Suite.eq(t_finish.Test_Suite).idxmax()

                    current_test_previous_status=test_suites_in_change_req.iat[test_suite_actual_index_currentTest-overlapping_row_count,3]
                    # get the status of currentTest in pre_req
                    t_finish_previous_status=test_suites_in_change_req.iat[test_suite_actual_index_tFinish-overlapping_row_count,3]
                    #  current_test failed and t_finish also failed
                    if(current_test_previous_status == "FAILED" and t_finish_previous_status == "FAILED"):
                        # increment the co fail count
                        co_fail_count+=1
                        
                    # if current test failed and t_finish passed
                    if(current_test_previous_status == "FAILED" and t_finish_previous_status == "PASSED"):
                        # increment the count of fail and pass together
                        co_fail_pass_count+=1

            except:
                print("Exception Occured")
                print("Shape of test suite in change req ",test_suites_in_change_req.shape)
                traceback.print_exc()
                pass
                
            
        #Assign conditional probability score  
        conditional_prob_score=0
     
        # if both currentTest and t_finish fails
        if t_finish.Status == "FAILED":
            if co_occurance_count != 0:
                conditional_prob_score=co_fail_count/co_occurance_count
        else:
            # if currentTest fail and t_finish pass
            if co_occurance_count != 0:
                conditional_prob_score=co_fail_pass_count/co_occurance_count
                
        if(conditional_prob_score!=0):
            new_score=previous_priority+(conditional_prob_score-0.5)
            # change the priority score to new_score
            finishedGpost.loc[finishedGpost['Test_Suite'] == currentTest.Test_Suite, 'Priority'] = new_score
            dispatch_queue.at[index, 'Priority'] = new_score
        # if the current test or previous test has status as failed, change the status as failed .
        # if the test is already present in history , a new row will not be added for that test suite again.
        # So update the status of existing row.
        
        if(previous_status == "FAILED" or currentTest.Status == "FAILED"):
            finishedGpost.loc[finishedGpost['Test_Suite'] == currentTest.Test_Suite, 'Status'] = "Failed"
        return True
    # if the test case is a new test case 
    else:
        return False
```

```python
def calculateProbabilityPassFail(status): 
    value=1
    if status=="PASSED":
        value=0.5
    return value
```

```python
def appendingToHistory(row,finishedGpost):
        priority=0.5
        finishedGpost=finishedGpost.append({'Test_Suite' : row.Test_Suite , 'Change_Request' : row.Change_Request, 
                                                'Stage':row.Stage, 'Status': row.Status, 'Launch_Time':row.Launch_Time, 
                                                      'Execution_Time':row.Execution_Time, 'Size':row.size, 'Shard_number':row.Shard_number,
                                                      'Run_number':row.Run_number, 'Language':row.Language,
                                              'Priority':priority}, ignore_index=True)
        
        return finishedGpost
```

```python
def addtoDispatchQueue(row,dispatch_queue):
        priority=0.5
        dispatch_queue =dispatch_queue.append({'Test_Suite' : row.Test_Suite , 'Change_Request' : row.Change_Request, 
                                                'Stage':row.Stage, 'Status': row.Status, 'Launch_Time':row.Launch_Time, 
                                                      'Execution_Time':row.Execution_Time, 'Size':row.size, 'Shard_number':row.Shard_number,
                                                      'Run_number':row.Run_number, 'Language':row.Language,
                                              'Priority':priority}, ignore_index=True)
        
        return dispatch_queue
```

```python
def codynaqsingle(googlepostdatas,finishedGpost,googlepost_rem,dispatch_queue):
    # iterate through waiting queue for reprioritization (row= test case to be reprioritized)
    for index, row in googlepost_rem.iterrows():
        # check if the previous test case and the current test case are from the same change request. 
        # Because the algorithm will be applied only if t1, t2 are in the same test request
        if googlepostdatas.loc[index, 'Change_Request'] == googlepostdatas.loc[index-1, 'Change_Request']:
            # get all the test cases from historical data set (newgooglepre) of current change request
            temp_test_suites_current_request= finishedGpost.loc[finishedGpost['Change_Request']==row.Change_Request] 
            # iterate through the historic test cases from the same change request for reprioritizing the current test case
            for index_j, row_j in temp_test_suites_current_request.iterrows():
                # temp suite can contain the same test case from historical result with same change request number
                #finishedGpost can contain the same test case from historical results with any change request
                result=reprioritize(row,index,row_j, finishedGpost,dispatch_queue)
                if(result==False):
                    finishedGpost=appendingToHistory(row,finishedGpost)
                    
        else:
            # append the test case to the historical data frame that has stores the tests that has finished execution
            finishedGpost=appendingToHistory(row,finishedGpost)
            
    return [finishedGpost,dispatch_queue]
    
```

### CodynaqDouble

```python
# create historical data frame= newgooglepre - eg. change req =5
# put them into waiting queue
# take certatin tests from the waiting queue and put them into dispatch queue
# reprioritize the dispatch queue continuously

codynaq_dispatchQ = pd.DataFrame()

def codynaqdouble(googlepostcommit, googlepost_finished,googlepost_rem):
    global codynaq_dispatchQ
    index=0
    size=googlepost_rem.Change_Request.count()
    print("Size : ",size)
    dispatchQueue=pd.DataFrame()
    while True :
        if(index<=size):
            print("Index<Size")
            if (len(dispatchQueue)==0):
                dispatchQueue=googlepost_rem[index:index+60]
                dispatch_queue = dispatchQueue.copy(deep = True)
                dispatch_queue['Priority'] = 0.5
                index=index+60
            if (len(dispatchQueue)!=0):
                result_data_frames = codynaqsingle(googlepostcommit, googlepost_finished,dispatchQueue,dispatch_queue)
                googlepost_finished = result_data_frames[0]
                temp_codynaqD = result_data_frames[1].sort_values('Priority', ascending = False)
                codynaq_dispatchQ = codynaq_dispatchQ.append(temp_codynaqD,ignore_index = True)                
                dispatchQueue=pd.DataFrame()
            print("Index : ",index)
        else:
            break        
    return codynaq_dispatchQ
```

```python
# FInding First fail
import pandas as pd
import numpy as np


def find_first_fail_all_fail(dispatch_queue,sortColumn,sortOrder,change_request):
    first_fail_sum=0
    all_fail_sum=0
    
    if(not (sortColumn=="None")):    
        dispatch_queue=dispatch_queue.sort_values(by = sortColumn,ascending = sortOrder)
    dispatch_queue_reindexed = dispatch_queue.reset_index()
    dispatch_queue_1 = dispatch_queue_reindexed[dispatch_queue_reindexed['Change_Request'] == change_request]
    dispatch_queue_failed = dispatch_queue_1[dispatch_queue_1['Status'] == "FAILED"]
    
    #First Fail,all fail
    if(len(dispatch_queue_failed)>0):
        first_fail_index  = dispatch_queue_failed.index[0]
        all_fail_index = dispatch_queue_failed.index[-1]
        first_fail_sum = dispatch_queue_reindexed[0:first_fail_index+1]['Execution_Time'].sum()
        all_fail_sum = dispatch_queue_reindexed[0:all_fail_index+1]['Execution_Time'].sum()              
    return [first_fail_sum,all_fail_sum,dispatch_queue_failed]

    
```
