### Reading and converting GooglePresCleanData and GooglePostCleanData files to csv files

```python
#Convert GooglePre dataset to csv files
import csv

with open('C:\Software testingpapers\SOEN-7481-Project-master\GooglePresCleanData.out', 'r') as in_file, open('C:\Software testingpapers\SOEN-7481-Project-master\GooglePresCleanData.csv', 'w') as out_file:
    stripped = (line.strip() for line in in_file)
    splited_lines = (line.split(",") for line in stripped)
    grouped_data = zip(splited_lines)
    writer = csv.writer(out_file)
    writer.writerow(("Test_Suite","Change_Request","Stage","Status","Launch_Time","Execution_Time","Size",
            "Shard_number","Run_number","Language"))
    for singleLine in grouped_data:
        writer.writerows(singleLine)
        
```

```python
# function for calculating pass fail rate
def CalculateRate(passFailCount):
    value=(passFailCount/(passFailCount.sum()))*100
    return value
```

## Success and Failure rate by change request number

```python
# This gives pass fail test status according to changerequest It uses google post data

postTotalPassFail_rate = googlepostcommit.groupby('Change_Request')['Status'].value_counts() 
postTotalPassFail_rate.groupby(level=0).apply(CalculateRate)  
```

```python
#Number of passed and failed tests by change request number for GooglePre Dataset

preTotalPassFail_rate = googleprecommit.groupby('Change_Request')['Status'].value_counts() 
preTotalPassFail_rate.groupby(level=0).apply(CalculateRate)  
```
